```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the dataset
df = pd.read_csv ('/home/ciline/Téléchargements/ship_fuel_efficiency.csv')
# --- 1. DATA INSPECTION ---
print("--- Dataset Head ---")
print(df.head())
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python

# --- 2. DESCRIPTIVE STATISTICS ---
print("\n--- Summary Statistics ---")
print(df.describe())
```

```python

# --- 3. PIE CHART: Ship Type Distribution ---
plt.figure(figsize=(8, 8))
df['ship_type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Distribution of Ship Types")
plt.ylabel('')
plt.show()
```

```python

# --- 4. COUNTPLOT: Weather Conditions ---
plt.figure(figsize=(10, 6))
sns.countplot(x='weather_conditions', data=df, palette='viridis')
plt.title('Frequency of Weather Conditions')
plt.show()
```

```python

# --- 5. HISTOGRAM: Engine Efficiency ---
plt.figure(figsize=(10, 6))
sns.histplot(df['engine_efficiency'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Engine Efficiency (%)')
plt.show()
```

```python

# --- 6. BAR CHART: Average Fuel Consumption ---
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='ship_type', y='fuel_consumption', palette='magma')
plt.title('Average Fuel Consumption by Ship Type')
plt.xticks(rotation=45)
plt.show()
```

```python

# --- 7. BOXPLOT: CO2 Emissions Outliers ---
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['CO2_emissions'], color='lightcoral')
plt.title('Outlier Detection in CO2 Emissions')
plt.show()
```

```python

# --- 8. HEATMAP: Correlation Matrix ---
plt.figure(figsize=(10, 8))
# We select only numeric columns for the correlation
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
